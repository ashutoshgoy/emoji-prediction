# Processing Data Sets

```python
import pandas as pd
import numpy as np
train=pd.read_csv('train_emoji.csv',header=None)
test=pd.read_csv('test_emoji.csv',header=None)
```

# Converting Sentence to Words

```python
embeddings = {}
with open('glove.6B.50d.txt',encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coeffs = np.asarray(values[1:],dtype='float32')
        
#         print(word)
#         print(coeffs)
        embeddings[word] = coeffs
```

```python
def getOutputEmbeddings(X):
    
    embedding_matrix_output = np.zeros((X.shape[0],10,50))
    for ix in range(X.shape[0]):
        X[ix] = X[ix].split()
        for jx in range(len(X[ix])):
            embedding_matrix_output[ix][jx] = embeddings[X[ix][jx].lower()]
            
    return embedding_matrix_output
```

```python
from keras.utils import to_categorical
from keras.layers import Dense,LSTM,Dropout
from keras.models import Sequential
```

```python
XT=train[0]
Xt=test[0]

YT=to_categorical(train[1])
Yt=to_categorical(test[1])
```

```python
emb_XT=getOutputEmbeddings(XT)
emb_Xt=getOutputEmbeddings(Xt)
```

```python
emb_XT
```

```python
emb_Xt
```

## Creating model

```python
model=Sequential()
model.add(LSTM(64,input_shape=(10,50)))
model.add(Dropout(0.5))
model.add(Dense(5,activation='softmax'))
model.summary()
```

```python

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['acc'])
hist=model.fit(emb_XT,YT,epochs=100,validation_split=0.2,shuffle=True,batch_size=64)
```

```python
pred = model.predict_classes(emb_Xt)
```

```python
for i in range(30):
    print(' '.join(Xt[i]))
    print(emoji.emojize(emoji_dictionary[str(np.argmax(Yt[i]))]))
    print(emoji.emojize(emoji_dictionary[str(pred[i])]))
```

```python
model.evaluate(emb_Xt,Yt)
```

```python
from keras.callbacks import EarlyStopping,ModelCheckpoint
earlystopping=EarlyStopping(patience=10,monitor='val_acc')
modelcheckpoint=ModelCheckpoint('my_model.h5',monitor='val_loss',verbose=True,save_best_only=True)

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['acc'])
hist=model.fit(emb_XT,YT,epochs=100,validation_split=0.2,shuffle=True,batch_size=64,callbacks=[modelcheckpoint,earlystopping])
```

```python
model.load_weights('my_model.h5')
```

```python
pred = model.predict_classes(emb_Xt)
for i in range(30):
    print(' '.join(Xt[i]))
    print(emoji.emojize(emoji_dictionary[str(np.argmax(Yt[i]))]))
    print(emoji.emojize(emoji_dictionary[str(pred[i])]))
```

```python
model.evaluate(emb_Xt,Yt)
```
